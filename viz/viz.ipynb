```python
import cv2
import matplotlib.pyplot as plt

# Membaca data gambar sebagai citra grayscale
img = cv2.imread("/Users/ramaido/gogo/ImProc/output_lena.png", cv2.IMREAD_GRAYSCALE)

# Menampilkan gambar
plt.imshow(img, cmap='gray')
plt.show()
```

```python
# Membaca data gambar
img_color = plt.imread("/Users/ramaido/gogo/ImProc/proceed_img/lena_gray.png")
img_gray = cv2.imread("/Users/ramaido/gogo/ImProc/output_lena.png", cv2.IMREAD_GRAYSCALE)

# Menampilkan gambar
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

# Gambar berwarna
axes[0].imshow(img_color, cmap='gray')
axes[0].set_title('Original Image')

# Gambar grayscale
axes[1].imshow(img_gray, cmap='gray')
axes[1].set_title('Equalized Image')

# Menampilkan histogram
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

# Histogram gambar berwarna
axes[0].hist(img_color.flatten(), bins=256, color='r')
axes[0].set_title('Histogram of Original Image')
axes[0].set_xlabel('Pixels Intentity')
axes[0].set_ylabel('Freq')

# Histogram gambar grayscale
axes[1].hist(img_gray.flatten(), bins=256, color='r')
axes[1].set_title('Histogram of Equalized Image')
axes[1].set_xlabel('Pixels Intentity')
axes[1].set_ylabel('Freq')

plt.tight_layout()
plt.show()
```

```python
# Membaca data gambar
img_color = plt.imread("/Users/ramaido/gogo/ImProc/proceed_img/lena_gray.png")
img_gray = cv2.imread("/Users/ramaido/gogo/ImProc/output_lena.png", cv2.IMREAD_GRAYSCALE)

# Menampilkan gambar
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

# Gambar berwarna
axes[0].imshow(img_color)
axes[0].set_title('Original Image (Color)')

# Gambar grayscale
axes[1].imshow(img_gray, cmap='gray')
axes[1].set_title('Equalized Image (Grayscale)')

# Menampilkan histogram
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

# Histogram gambar berwarna
for i in range(3):  # Loop untuk tiga saluran warna (R, G, B)
    axes[0].hist(img_color[:, :, i].flatten(), bins=256, color=['r', 'g', 'b'][i], alpha=0.5, label=['R', 'G', 'B'][i])

axes[0].set_title('Histogram of Original Image (Color)')
axes[0].set_xlabel('Pixels Intensity')
axes[0].set_ylabel('Freq')
axes[0].legend()

# Histogram gambar grayscale
axes[1].hist(img_gray.flatten(), bins=256, color='gray', alpha=0.5)
axes[1].set_title('Histogram of Equalized Image (Grayscale)')
axes[1].set_xlabel('Pixels Intensity')
axes[1].set_ylabel('Freq')

plt.tight_layout()
plt.show()
```

```python
# Membaca data gambar
img_color = cv2.imread("/Users/ramaido/gogo/ImProc/proceed_img/lena_gray.png")
img_gray = cv2.imread("/Users/ramaido/gogo/ImProc/output_lena.png", cv2.IMREAD_GRAYSCALE)

# Menampilkan gambar
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

# Gambar berwarna
axes[0].imshow(cv2.cvtColor(img_color, cv2.COLOR_BGR2RGB))  # Perlu diubah ke format RGB untuk tampilan yang benar
axes[0].set_title('Original Image')

# Gambar grayscale
axes[1].imshow(img_gray, cmap='gray')
axes[1].set_title('Equalized Image')

# Menampilkan histogram
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

# Histogram gambar berwarna
axes[0].hist(img_color.flatten(), bins=256, color='r')
axes[0].set_title('Histogram of Original Image')
axes[0].set_xlabel('Pixels Intensity')
axes[0].set_ylabel('Freq')

# Histogram gambar grayscale
axes[1].hist(img_gray.flatten(), bins=256, color='r')
axes[1].set_title('Histogram of Equalized Image')
axes[1].set_xlabel('Pixels Intensity')
axes[1].set_ylabel('Freq')

plt.tight_layout()
plt.show()
```
